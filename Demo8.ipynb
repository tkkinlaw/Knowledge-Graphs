```python
import arcpy
import os
import ast
```

```python
file = r"C:\Users\tim10393\Downloads\dataMetals.txt"
```

```python
arcpy.env.overwriteOutput = True
tableDesc = arcpy.da.Describe(file)
gdb = "Scratch.gdb"
table = "SuppliersAndDistributorsMetals"
arcpy.management.CreateFileGDB(tableDesc['path'], gdb)
arcpy.env.workspace = os.path.join(tableDesc['path'], gdb)
arcpy.management.CreateTable(out_path=arcpy.env.workspace, out_name=table)
arcpy.management.AddFields(in_table=table, field_description=[['OriginCountry', 'TEXT'],['Supplier', 'TEXT'],['DestinationCountry', 'TEXT'],['Distributor', 'TEXT']])
```

```python
data = open(file)
cursor = arcpy.da.InsertCursor(table, ['OriginCountry', 'Supplier', 'DestinationCountry', 'Distributor'])

for record in data:
    row = ast.literal_eval(record.strip('\n'))
    print(row)
    cursor.insertRow(row)
data.close()
del cursor
```
